```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el DataFrame desde el archivo CSV
df_merged = pd.read_csv('df_merged.csv')

# Contar el número de juegos por género
genre_counts = df_merged['genres'].value_counts()

# Filtrar géneros que cubran el 80% de los juegos
total_juegos = len(df_merged)
genre_counts = genre_counts[genre_counts.cumsum() <= 0.8*total_juegos]

# Crear un gráfico de barras
plt.figure(figsize=(10, 6))
genre_counts.plot(kind='bar', color='salmon')
plt.title('Número de Juegos por Género (Top 80%)')
plt.xlabel('Género')
plt.ylabel('Número de Juegos')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el DataFrame desde el archivo CSV
df_merged = pd.read_csv('df_merged.csv')

# Ordenar los títulos por tiempo de uso en orden descendente
df_sorted = df_merged.sort_values(by='playtime_forever', ascending=False)

# Calcular el tiempo total
total_playtime = df_sorted['playtime_forever'].sum()

# Calcular el 80% del tiempo total
threshold_playtime = 0.8 * total_playtime

# Calcular el tiempo acumulado y seleccionar los títulos
df_sorted['cumulative_playtime'] = df_sorted['playtime_forever'].cumsum()
df_filtered = df_sorted[df_sorted['cumulative_playtime'] <= threshold_playtime]

# Crear un gráfico de dispersión para playtime_forever vs title
plt.figure(figsize=(10, 6))
plt.scatter(df_filtered['title'], df_filtered['playtime_forever'], alpha=0.5, color='red')
plt.title('Tiempo de Uso vs Título del Videojuego (80% del tiempo total)')
plt.xlabel('Título del Videojuego')
plt.ylabel('Tiempo de Uso')
plt.xticks(rotation=90)
plt.show()
```
